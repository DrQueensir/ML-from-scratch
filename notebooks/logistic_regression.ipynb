```python
import time

import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression as SKLogisticRegression

from sklearn.metrics import accuracy_score
from algorithms.linear_models.logistic_regression import LogisticRegression
```

```python
X, y = make_classification(
    n_samples=1000,
    n_features=2,
    n_redundant=0,
    n_informative=2,
    n_clusters_per_class=1,
    random_state=42
)

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
plt.scatter(
    X[:,0],
    X[:,1],
)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```

```python
model = LogisticRegression()

start = time.perf_counter()

model.fit(X,y)

training_time = time.perf_counter() - start

predictions = model.predict(X)
```

```python
plt.plot(model.cost_history)

plt.xlabel("Iteration")
plt.ylabel("Cross Entropy Loss")

plt.show()
```

```python
sk_model = SKLogisticRegression()

start = time.perf_counter()

sk_model.fit(X,y)

sk_time = time.perf_counter()-start

sk_predictions = sk_model.predict(X)
```

```python
print("Your Model :", accuracy_score(y,predictions))

print("Scikit :", accuracy_score(y,sk_predictions))
```

```python
plt.bar(
    ["Your Model","Scikit"],
    [training_time,sk_time]
)

plt.ylabel("Seconds")

plt.show()
```
